```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/robyn_dt_simulated.csv", parse_dates=["DATE"])

# Fix 1: 'events' column has "na" strings — convert to proper NaN then encode
df["events"] = df["events"].replace("na", np.nan)
df["has_event"] = df["events"].notna().astype(int)

# Fix 2: rename to match pipeline.py
df = df.rename(columns={"DATE": "date", "revenue": "revenue"})

print(df.shape)
print(df["tv_S"].describe())
print(df["date"].min(), df["date"].max())
```

```python
df.to_csv("data/robyn_dt_simulated.csv", index=False)
```
